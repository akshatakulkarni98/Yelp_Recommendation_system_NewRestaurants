```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Data/restaurant_review.csv',index_col=0)
df
```

```python
business_id_and_category_dict = dict()
categories_set = set()

for index, row in df.iterrows():
    #print(row['name'], row['categories'])
    business_id_and_category_dict[row['business_id']] = row['categories']
    for category in row['categories'].split(','):
        categories_set.add(category)

print (len(categories_set))
```

```python
print (sorted(categories_set))
```

```python
df_business_id_only = df [['business_id']]
df_business_id_only = df_business_id_only.drop_duplicates(inplace=False)
df_business_id_only = df_business_id_only.set_index('business_id')
df_business_id_only
```

```python
categories_list = sorted(categories_set)
df_final = pd.DataFrame(categories_list)
df_final = df_final.transpose()
header = df_final.iloc[0]
```

```python

df_final = df_final[1:]
df_final = df_final.rename(columns = header)
df_final
```

```python
df_final = pd.concat([df_business_id_only, df_final], axis = 1)
```

```python
df_final
```

```python
business_id_and_category_dict
```

```python
df_final = df_final.replace(np.NaN,0)
df_final.loc['f-m7-hyFzkf0HSEeQ2s-9A']['Afghan']
```

```python
for business_id,categories in business_id_and_category_dict.items():
    for category in categories.split(','):
        df_final.loc[business_id][category]=1
```
